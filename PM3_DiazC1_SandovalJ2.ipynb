## Sistema masa-resorte: movimiento libre amortiguado

### 1.2 Objetivo específico

## 2. Modelo que representa el problema

El concepto de movimiento armónico libre es un poco irreal, puesto que el movimiento de dicha ecuación supone que no hay fuerzas retardadoras actuando sobre la masa en movimiento.

La solución a este sistema masa-resorte amortiguado se explica en términos de la segunda ley de Newton. Para este caso, si la masa permanece constante y solo consideramos la dirección en $x$. Entonces,
\begin{equation}
F = m \frac{d^2x}{dt^2}.
\end{equation}

La fuerza se define como:
\begin{equation}
F = -k x - \beta\frac{dx}{dt}, \quad k > 0.
\end{equation}

Donde $\beta$ es una constante de amortiguamiento positiva y el signo negativo es una consecuencia del hecho de que la fuerza de amortiguamiento actúa en una dirección opuesta al movimiento.

Dividiendo la ecuación entre la masa m se encuentra que la ecuación diferencial del movimiento libre amortiguado es:

\begin{equation}
\frac{d^2x}{dt^2} +2\lambda \frac{dx}{dt}+\omega^2x = 0,
\end{equation}

Donde:

\begin{equation}
2\lambda=\frac{\beta}{m}, \omega^2=\frac{k}{m} 
\end{equation}

El símbolo $2\lambda$ se usa solo por conveniencia algebraica, porque la ecuación auxiliar es $m^2+2(\lambda)m+\omega^2=0$ 

$$m_1=-\lambda+\sqrt{\lambda^2-\omega^2}, m_2=-\lambda-\sqrt{\lambda^2-\omega^2}$$

Con ello se pueden distinguir tres casos posibles dependiendo del signo algebraico de $\lambda^2-\omega^2$. Puesto que cada solución contiene el factor de amortiguamiento $e^{-(\lambda)t}, \lambda>0$, los desplazamientos de la masa se vuelven despreciables conforme el tiempo $t$ aumenta.

Para el primer caso $\lambda^2-\omega^2>0$. En esta situación el sistema se conoce como sobreamortiguado porque el coeficiente de amortiguamiento $\beta$ es grande comparado con la constante del resorte $k$.

### 2.1 Problema

Resuelva la ecuación diferencial para los tres casos:

1. $\lambda^2 = \omega ^2$, con $m=1$, $\beta$=2 y $k=1$
2. $\lambda^2 > \omega ^2$, con $m=1$, $\beta=4$ y $k=1$
3. $\lambda^2 < \omega ^2$, con $m=1$, $\beta=0.5$ y $k=2$


## 3. Solución

En la ecuación diferencial

$$\frac{d^2x}{dt^2} +2\lambda \frac{dx}{dt}+\omega^2x = 0$$

hacemos el siguiente cambio de variable:

$$x_1=x$$
$$x_2=\frac{dx}{dt}=\frac{dx_1}{dt} \rightarrow \frac{dx_2}{dt}=\frac{d^2x}{dt^2}$$

Y tenemos el siguiente sistema de ecuaciones diferenciales:

$$\frac{dx_1}{dt}=x_2$$
$$\frac{dx_2}{dt}=-2\lambda x_2-\omega ^2x_1$$

Para el primer caso $\lambda^2-\omega^2>0$. En esta situación el sistema sobreamortiguado porque el coeficiente de amortiguamiento $\beta$ es grande comparado con la constante del resorte $k$.

### 3.1 $\lambda^2=\omega^2$

```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
from scipy.integrate import odeint
```

```python
beta=2
k=1
m=1
lamb=beta/(2*m)
om=k/m
def eqs(x,t):
    x1=x[0]
    x2=x[1]
    dx1=x2
    dx2=-2*lamb*x2-(om**2)*x1
    return np.array([dx1,dx2])
```

```python
x0 = np.array([0,3])
t = np.linspace(0,25,300)
```

```python
xx = odeint(eqs, x0, t)
x1 = xx[:, 0]
x2 = xx[:, 1]

plt.figure(figsize=(8,8))
plt.plot(t, x1, label='$x(t)$')
plt.plot(t, x2, label='$\dot{x}(t)$')
plt.xlabel('Tiempo $t$', fontsize=12)
plt.axhline(y=0, color='k')
plt.axvline(x=0, color='k')
plt.title('Posición y velocidad en el tiempo',fontsize=18)
plt.legend(loc='best')
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(8,8))
plt.plot(x1, x2, label='Trayectoria')
plt.plot(x1[0], x2[0], '*', ms=10, label = 'Trayectoria inicial')
plt.plot(x1[-1], x2[-1], '*', ms=10, label = 'Trayectoria final')
plt.xlabel('$x_1$')
plt.ylabel('$x_2$')
plt.axhline(y=0, color='k')
plt.axvline(x=0, color='k')
plt.title('Espacio fase',fontsize=18)
plt.legend(loc = 'best')
plt.grid()
plt.show()
```
