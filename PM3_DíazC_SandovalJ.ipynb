## Sistema masa-resorte: movimiento libre amortiguado

El concepto de movimiento armónico libre es un poco irreal, puesto que el movimiento de dicha ecuación supone que no hay fuerzas retardadoras actuando sobre la masa en movimiento.

La solución a este sistema masa-resorte amortiguado se explica en términos de la segunda ley de Newton. Para este caso, si la masa permanece constante y solo consideramos la dirección en $x$. Entonces,
\begin{equation}
F = m \frac{d^2x}{dt^2}.
\end{equation}

La fuerza se define como:
\begin{equation}
F = -k x - \beta\frac{dx}{dt}, \quad k > 0.
\end{equation}

Donde \beta es una constante de amortiguamiento positiva y el signo negativo es una consecuencia del hecho de que la fuerza de amortiguamiento actúa en una dirección opuesta al movimiento.

Dividiendo la ecuación entre la masa m se encuentra que la ecuación diferencial del movimiento libre amortiguado es:

\begin{equation}
\frac{d^2x}{dt^2} +2\lambda \frac{dx}{dt}+\omega^2x = 0,
\end{equation}

Donde:

\begin{equation}
2\lambda=\frac{\beta}{m}, \omega^2=\frac{k}{m} 
\end{equation}

El símbolo $2\lambda$ se usa solo por conveniencia algebraica, porque la ecuación auxiliar es $m^2+2(\lambda)m+\omega^2=0$ 

$$m_1=-\lambda+\sqrt{\lambda^2-\omega^2}, m_2=-\lambda-\sqrt{\lambda^2-\omega^2}$$

Con ello se pueden distinguir tres casos posibles dependiendo del signo algebraico de $\lambda^2-\omega^2$. Puesto que cada solución contiene el factor de amortiguamiento $e^(-(\lambda)t), \lambda>0$, los desplazamientos de la masa se vuelven despreciables conforme el tiempo $t$ aumenta.

Para el primer caso $\lambda^2-\omega^2>0$. En esta situación el sistema sobreamortiguado porque el coeficiente de amortiguamiento $\beta$ es grande comparado con la constante del resorte $k$. La solución es:
\begin{equation}
x(t) = e^{-\lambda{t}}(Ae^{\sqrt{\lambda^2-\omega^2}t}+Be^{-\sqrt{\lambda^2-\omega^2}t})
\end{equation}
Y su primera derivada (velocidad) sería 
\begin{equation}
v(t) = \frac{dx(t)}{dt} = e^{-t\sqrt{\lambda^2-\omega^2}+\lambda}(A\sqrt{\lambda^2-\omega^2}-\lambda)e^{2t\sqrt{\lambda^2-\omega^2}}-B\sqrt{\lambda^2-\omega^2}+\lambda)
\end{equation}

Para el segundo caso $\lambda^2-\omega^2=0$. Este sistema está criticamente amortiguado porque cualquier ligera disminución en la fuerza de amortiguamiento daría como resultado un movimiento oscilatorio.

```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
from scipy.integrate import odeint
```

```python
def funcion(x, t):
    l=1
    w=1
    x1 = x[0]
    x2 = x[1]
    dx1 = x2
    dx2 = -2*l*x2-(w**2)*x1
    return np.array([dx1, dx2])
```

```python
x0 = np.array([0,3])
t = np.linspace(0,25,300)
```

```python
xx = odeint(funcion, x0, t)
x1 = xx[:, 0]
x2 = xx[:, 1]
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
plt.figure(figsize=(8,6))
plt.plot(t, x1, label='$x_1(t)$')
plt.plot(t, x2, label='$x_2(t)$')
plt.xlabel('Tiempo $t$')
plt.axhline(y=0, color='k')
plt.axvline(x=0, color='k')
plt.legend(loc='best')
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(8,6))
plt.plot(x1, x2, label='Trayectoria')
plt.plot(x1[0], x2[0], '*', ms=10, label = 'Trayectoria inicial')
plt.plot(x1[-1], x2[-1], '*', ms=10, label = 'Trayectoria final')
plt.xlabel('$x_1$')
plt.ylabel('$x_2$')
plt.axhline(y=0, color='k')
plt.axvline(x=0, color='k')
plt.legend(loc = 'best')
plt.grid()
plt.show()
```
